```python
number_student_df = df.copy()
```

```python
number_student_df['NUMBER_OF_STUDENT'] = number_student_df.groupby(['COURSECODE','COURSENAMEENG'])['COURSECODE'].transform('count')
```

```python
grade_df_list = number_student_df.groupby(['COURSECODE','COURSENAMEENG','NUMBER_OF_STUDENT'])['GRADE'].apply(list).reset_index()
```

```python
grade_df_list
```

```python
grade_df = pd.DataFrame(grade_df_list)
```

```python
grade_df.head()
```

```python
grade_df.columns
```

```python
def count_grade(grade_col,grade_value):
    grade_count = grade_col.count(grade_value)
    return grade_count
```

```python
for grade in grade_list:
    grade_df[grade] = grade_df["GRADE"].apply(count_grade,grade_value=grade)
```
